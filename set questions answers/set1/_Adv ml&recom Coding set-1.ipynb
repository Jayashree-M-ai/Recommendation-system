```python
#1.
import pandas as pd

df = pd.DataFrame({
    'Age': [25, 32, 47, 51, 62],
    'Salary': [40000, 50000, 60000, 52000, 62000],
    'Purchased': [0, 1, 1, 0, 1]
})

X = df[['Age', 'Salary']]
y = df['Purchased']
Use Mutual Information to find the most important feature for predicting Purchased. Select the top 1 feature.

Expected Output:
['Salary']
```

```python
import pandas as pd
from sklearn.feature_selection import SelectKBest, mutual_info_classif

df = pd.DataFrame({
    'Age': [25, 32, 47, 51, 62],
    'Salary': [40000, 50000, 60000, 52000, 62000],
    'Purchased': [0, 1, 1, 0, 1]
})

X = df[['Age', 'Salary']]
y = df['Purchased']

# Select top 1 feature using Mutual Information
selector = SelectKBest(score_func=mutual_info_classif, k=1)
selector.fit(X, y)

selected_feature = X.columns[selector.get_support()]
print(list(selected_feature))
```

```python
#2.
import pandas as pd

data = pd.DataFrame({
    'User': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Item': ['Apple', 'Banana', 'Apple', 'Orange', 'Banana', 'Orange'],
    'Rating': [5, 3, 4, 2, 5, 4]
})

Based on item popularity (sum of ratings), recommend top 2 items to a new user.

 Expected Output:

['Apple', 'Banana']

```

```python
import pandas as pd

data = pd.DataFrame({
    'User': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Item': ['Apple', 'Banana', 'Apple', 'Orange', 'Banana', 'Orange'],
    'Rating': [5, 3, 4, 2, 5, 4]
})

# Compute popularity = sum of ratings per item
popular_items = data.groupby('Item')['Rating'].sum()

print(popular_items)

# Get top 2 items
top_2 = popular_items.sort_values(ascending=False).head(2).index.tolist()

print("Top 2 items to be recommeded to the new user are",top_2)
    
```

```python
#3.
from sklearn.datasets import load_diabetes
import pandas as pd

X, y = load_diabetes(return_X_y=True, as_frame=True)
Write a Python program to perform feature selection using Recursive Feature Elimination (RFE). Use Linear Regression as the estimator and select the top 5 features. Print the selected feature names.

Expected output:
Selected features: ['bmi', 's5', 'bp', 'age', 'sex']
```

```python
from sklearn.datasets import load_diabetes
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

# Load dataset
X, y = load_diabetes(return_X_y=True, as_frame=True)

print("X Features: ",X.columns)

# Create the model
model = LinearRegression()

# Apply RFE to select top 5 features
rfe = RFE(estimator=model, n_features_to_select=5)
rfe.fit(X, y)

# Get selected feature names
selected_features = X.columns[rfe.support_]

print("Selected features:", list(selected_features))
```

```python
#5.
from sklearn.datasets import load_breast_cancer
import pandas as pd

X, y = load_breast_cancer(return_X_y=True, as_frame=True)
Write a Python program to perform feature selection using SelectFromModel. Train a Logistic Regression model and select features based on importance using SelectFromModel. Print the names of the selected feature


Expected Output:
Selected features: ['mean radius', 'mean concavity', 'texture error', 'worst texture', 'worst smoothness', 'worst compactness', 'worst concavity', 'worst concave points', 'worst symmetry']
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectFromModel

# Load data
X, y = load_breast_cancer(return_X_y=True, as_frame=True)

print("X Features: ",X.columns)

# Train Logistic Regression
model = LogisticRegression(max_iter=5000)
model.fit(X, y)

# Select important features
selector = SelectFromModel(model, prefit=True)
selected_features = X.columns[selector.get_support()]

print("Selected features:", list(selected_features))
```
