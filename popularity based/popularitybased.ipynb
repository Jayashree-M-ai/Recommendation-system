```python
# importing libraries
import pandas as pd
import numpy as np
```

```python
# load datasets
i_cols = ['movie_id', 'movie_title' ,'release_date','video_release_date', 'IMDb_URL', 'unknown', 'Action', 'Adventure',
'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
movies = pd.read_csv('u.item', sep='|', names=i_cols,
encoding='latin-1')

ratings = pd.read_csv("u.data", sep="\t",
                      names=["user_id","movie_id","rating","timestamp"])
```

```python
movies
```

```python
ratings
```

```python
# compute rating stats
movie_stats = ratings.groupby("movie_id").agg(
    avg_rating=("rating","mean"),
    num_ratings=("rating","count")
).reset_index()
```

```python
movie_stats
```

```python
# merge
df = movies.merge(movie_stats, on="movie_id")
```

```python
df
```

```python
# filter movies with at least 50 ratings
df = df[df["num_ratings"] >= 50]
```

```python
df
```

```python
# list of genre column indexes (last 19 columns)
genre_cols = movies.columns[-18:]
```

```python
genre_cols
```

```python
# Print top 5 movie names under each genre
for genre in genre_cols:
    movies_in_genre = df[df[genre] == 1].sort_values(
        by=["avg_rating","num_ratings"], ascending=False
    ).head(5)

    print(f"\n=== {genre} ===")
    for title in movies_in_genre['movie_title']:
        print(title)
```
